```python
def classify_sample(sample_name):
    sample_name = str(sample_name)
    if 'GC' in sample_name:
        return 'GC'
    elif 'VIV' in sample_name:
        return 'VIV'
    elif 'FLT' in sample_name:
        return 'FLT'
    elif 'BSL' in sample_name:
        return 'BSL'
    else:
        return 'Unknown'

merged_df['mice_type'] = merged_df['Mice'].map(classify_sample)
```

```python
def classify_sample_2(sample_name):
    sample_name = str(sample_name)
    if 'ISS-T' in sample_name:
        return 'ISS'
    elif 'LAR' in sample_name:
        return 'LAR'
    else:
        return 'Unknown'

merged_df['Place'] = merged_df['Mice'].map(classify_sample_2)
```

```python
def classify_sample_3(sample_name):
    sample_name = str(sample_name)
    if 'OLD' in sample_name:
        return 'Old'
    elif 'YNG' in sample_name:
        return 'Young'
    else:
        return 'Unknown'

merged_df['Age'] = merged_df['Mice'].map(classify_sample_3)
```

```python
y_axis_limits = {
    'total_sequences': (48000000, 83000000),
    'total_deduplicated_percentage': (9, 48),
    'percent_duplicates': (51, 91),
}

color_map = {
    'FLT': '#f9e79f',
    'GC': '#a9cce3',
    'VIV': '#99a3a4',
}


app = dash.Dash(__name__)


app.layout = html.Div([
    dcc.Dropdown(
        id='y-axis-dropdown',
        options=[{'label': col, 'value': col} for col in y_axis_limits.keys()],
        value='total_sequences',
        clearable=False,
        placeholder='Select Y Axis'
    ),
    dcc.Dropdown(
        id='mice-type-checklist',
        options=[{'label': mice_type, 'value': mice_type} for mice_type in merged_df['mice_type'].unique()],
        value=merged_df['mice_type'].unique().tolist(),
        multi=True,
        placeholder='Select Habitat type'
    ),
    dcc.Dropdown(
        id='place-checklist',
        options=[{'label': place, 'value': place} for place in merged_df['Place'].unique()],
        value=merged_df['Place'].unique().tolist(),
        multi=True,
        placeholder='Select Place'
    ),
    dcc.Dropdown(
        id='age-checklist',
        options=[{'label': age, 'value': age} for age in merged_df['Age'].unique()],
        value=merged_df['Age'].unique().tolist(),
        multi=True,
        placeholder='Select Age'
    ),
    dcc.Graph(id='bar-plot')
])


@app.callback(
    Output('bar-plot', 'figure'),
    Input('y-axis-dropdown', 'value'),
    Input('mice-type-checklist', 'value'),
    Input('place-checklist', 'value'),
    Input('age-checklist', 'value')
)
def update_plot(y_axis, mice_types, places, ages):
    filtered_df = merged_df[
        (merged_df['mice_type'].isin(mice_types)) &
        (merged_df['Place'].isin(places)) &
        (merged_df['Age'].isin(ages))
    ]

    if filtered_df.empty:
        return px.bar()  

    
    sorted_df = filtered_df.sort_values(by='Mice')

    fig = px.bar(
        sorted_df,
        x='Mice',
        y=y_axis,
        title=f"Bar Plot of {y_axis} by Mice",
        labels={y_axis: y_axis, 'Mice': 'Mice'},
        height=800,
        color='mice_type',
        color_discrete_map=color_map
    )

    
    fig.update_layout(
        xaxis_title="Mice Names",
        yaxis_title=y_axis,
        hovermode="x unified",
        plot_bgcolor='rgba(0,0,0,0)',
        yaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgrey')
    )

   
    y_limits = y_axis_limits[y_axis]
    fig.update_yaxes(range=y_limits)

    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8059)
```

```python
merged_df['percent_gc'] = merged_df['percent_gc'].astype(str)
merged_df['percent_fails']=merged_df['percent_fails'].astype(str)  
```

```python
columns_to_plot = ['per_tile_sequence_quality', 'per_base_sequence_content', 
                   'per_sequence_gc_content', 'adapter_content', 'percent_gc', 'percent_fails']

y_axis_limits = {
    'per_tile_sequence_quality': (0, 40),  
    'per_base_sequence_content': (0, 55),
    'per_sequence_gc_content': (0, 48),
    'adapter_content': (0, 40),
    'percent_gc': (0, 24),
    'percent_fails': (0, 32),
}


app = Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='y-axis-dropdown',
        options=[{'label': col, 'value': col} for col in columns_to_plot],
        value=columns_to_plot[0], 
        clearable=False,
        style={'width': '50%'}
    ),

    dcc.Dropdown(
        id='mice-type-checklist',
        options=[{'label': mice_type, 'value': mice_type} for mice_type in merged_df['mice_type'].unique()],
        value=merged_df['mice_type'].unique().tolist(),  
        multi=True,
        placeholder='Select Habitat type'
    ),
    dcc.Dropdown(
        id='place-checklist',
        options=[{'label': place, 'value': place} for place in merged_df['Place'].unique()],
        value=merged_df['Place'].unique().tolist(),  
        multi=True,
        placeholder='Select Place'
    ),
    dcc.Dropdown(
        id='age-checklist',
        options=[{'label': age, 'value': age} for age in merged_df['Age'].unique()],
        value=merged_df['Age'].unique().tolist(),  
        multi=True,
        placeholder='Select Age'
    ),
    dcc.Graph(id='bar-plot')
])


@app.callback(
    Output('bar-plot', 'figure'),
    Input('y-axis-dropdown', 'value'),
    Input('mice-type-checklist', 'value'),
    Input('place-checklist', 'value'),
    Input('age-checklist', 'value')
)
def update_plot(selected_column, selected_mice_types, selected_places, selected_ages):
   
    filtered_df = merged_df[
        (merged_df['mice_type'].isin(selected_mice_types)) &
        (merged_df['Place'].isin(selected_places)) &
        (merged_df['Age'].isin(selected_ages))
    ]

    frequency_df = filtered_df.groupby(['mice_type', selected_column]).size().reset_index(name='Frequency')
    unique_values = filtered_df[selected_column].unique()
    color_numbers = ['#f9e79f', '#a9cce3', '#99a3a4', '#f5b7b1', '#76d7c4', '#ffcc00']
    color_map = {value: color_numbers[i % len(color_numbers)] for i, value in enumerate(unique_values)}

    
    fig = px.bar(
        frequency_df,
        x='mice_type',
        y='Frequency',
        color=selected_column,
        title=f"Frequency of {selected_column} by Mice Type",
        labels={'Frequency': 'Count', 'mice_type': 'Mice Type'},
        height=600,
        color_discrete_map=color_map
    )

    
    fig.update_layout(
        barmode='group',
        xaxis_title='Mice Type',
        yaxis_title='Count',
        plot_bgcolor='rgba(0,0,0,0)',
        yaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgrey')
    )

    
    y_limits = y_axis_limits[selected_column]
    fig.update_yaxes(range=y_limits)

    return fig


if __name__ == '__main__':
    app.run_server(debug=True, port=8060)
```

```python
columns_to_plot = ['total_sequences', 'total_deduplicated_percentage', 'percent_duplicates']


y_axis_limits = {
    'total_sequences': (50000000, 65000000),
    'total_deduplicated_percentage': (10, 30),
    'percent_duplicates': (50, 75),
}


app = dash.Dash(__name__)


app.layout = html.Div([
    dcc.Dropdown(
        id='y-axis-dropdown',
        options=[{'label': col, 'value': col} for col in columns_to_plot],
        value=columns_to_plot[0],
        clearable=False,
        placeholder='Select Y Axis'
    ),
    dcc.Dropdown(
        id='x-axis-dropdown',  
        options=[{'label': 'Mice Type', 'value': 'mice_type'},
                 {'label': 'Place', 'value': 'Place'},
                 {'label': 'Age', 'value': 'Age'}],
        value='mice_type',  
        clearable=False,    
        placeholder='Select X Axis',
        multi=False         
    ),
    dcc.Graph(id='mean-bar-plot')
])

@app.callback(
    Output('mean-bar-plot', 'figure'),
    Input('y-axis-dropdown', 'value'),
    Input('x-axis-dropdown', 'value')  
)
def update_plot(selected_column, selected_x_category):   
    mean_df = merged_df.groupby(selected_x_category)[selected_column].mean().reset_index()
    mean_df.columns = [selected_x_category, selected_column]  

    
    unique_values = merged_df[selected_x_category].unique()
    color_map = {value: ['#f9e79f', '#a9cce3', '#99a3a4', '#f5b7b1'][i % 4] for i, value in enumerate(unique_values)}

    fig = px.bar(
        mean_df,
        x=selected_x_category,  
        y=selected_column,            
        title=f"Mean {selected_column} by {selected_x_category}",
        labels={selected_x_category: selected_x_category, selected_column: selected_column},
        height=600,
        color=selected_x_category,  
        color_discrete_map=color_map  
    )

    
    fig.update_layout(
        xaxis_title=selected_x_category,
        yaxis_title='Mean Value',
        plot_bgcolor='rgba(0,0,0,0)',
        yaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgrey')
    )

    
    y_limits = y_axis_limits[selected_column]
    fig.update_yaxes(range=y_limits)

    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8061)
```
